Standarizing the numeric fields

```
import pandas as pd
url = 'https://raw.githubusercontent.com/ArashVafa/DESC624/master/bank_marketing_training'
bank_train = pd.read_csv(url)
bank_train.head(10)
```

Below appears to add a new column "age_z" into the dataset bank_train.  The new column uses the stats.zscore to standardize the 'age' column using zscoring.

```
from scipy import stats
bank_train['age_z'] = stats.zscore(bank_train['age'])
```

```
bank_train['age'].plot(kind = 'hist')
```

```
bank_train['age_z'].plot(kind = 'hist')
```

**IDENTIFYING OUTLIERS**

Below appears to query the bank_train results, returning anything greater than 3 or less than -3 and place the results into the *bank_train_outliers* variable, but not display them.

```
bank_train_outliers = bank_train.query('age_z > 3' or 'age_z < -3')
```
